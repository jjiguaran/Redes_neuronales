```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
```

## AND

```python
Clase_A_x = [0 , 1, 0, 1]
Clase_A_y = [0, 0, 1, 1]
Clase_A_l = [0, 0, 0, 1]

clase_A = pd.DataFrame(
    {'x': Clase_A_x,
     'y': Clase_A_y,
     'label': Clase_A_l
    })
clase_A
```

```python
clases = clase_A
# clases['label'] = clases['label'].astype(str)
```

```python
X = clases[['x','y']].to_numpy()
X
```

```python
Y = clases[['label']].to_numpy()
Y
```

```python
P = X
T = Y
alfa = 0.3
max_itera = 10

(cant_patrones, cant_atrib) = P.shape
W = np.random.rand(1, cant_atrib)

W = np.array([0, 0.25])

b = np.random.rand()
b = -1.5
ite = 1
otra_vez = True

plt.ion()
plt.show()

while ((ite <= max_itera) and otra_vez):
    print('\n***Iteration: ', ite)
    otra_vez = False
    ite = ite + 1

    for patr in range(cant_patrones):
        print('\n Obs', patr)
        print(W.dot(P[patr, :][np.newaxis].T) )
        salida = b + W.dot(P[patr, :][np.newaxis].T) 
        print('Salida: ', salida)
        if salida >= 0:
            salida = 1
        else:
            salida = 0
        
        print(salida)
        
        factor = alfa * (T[patr] - salida)
        print('Previous: ', W)
        if (factor != 0):
            otra_vez = True
            W = W + factor * P[patr, :][np.newaxis]
            print('New W: ', W)
#             b = b + factor
W
```
