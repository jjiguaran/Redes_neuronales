# Entrena un modelo de Redes Neuronales

```python
import tensorflow as tf
import keras
import numpy as np
import utils 
import os

# Selección del dataset; descomenta el que quieras usar
dataset = "6_clases_dificil.csv"

# Carga del dataset
dataset_path=os.path.join("datasets_clasificacion",dataset)
data=np.loadtxt(open(dataset_path, "rb"), delimiter=",", skiprows=1)
x,y=data[:,0:-1],data[:,-1]
# cantidad de ejemplos y dimension de entrada
n,d_in=x.shape
# calcula la cantidad de clases
classes=int(y.max()+1)

print("Información del conjunto de datos:")
print(f"Ejemplos: {n}")
print(f"Variables de entrada: {d_in}")
print(f"Cantidad de clases: {classes}")

# Normalizo las variables de entrada
#for i in range(d_in):
#    x[:,i]=(x[:,i]-x[:,i].mean())/x[:,i].std()
```

```python
# Separacion en train y test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
# Normalizo las variables de entrada
for i in range(d_in):
    x_train[:,i]=(x_train[:,i]-x_train[:,i].mean())/x_train[:,i].std()
```

```python
# Creación del modelo inicial
print("Inicialización aleatoria del modelo (podes volver a correr esta celda para obtener otros resultados)")

# Creo un modelo Red Neuronal 
modelo = keras.Sequential([
    # input_shape solo en la primer capa
    # Capa con 3 salidas, activación relu
    keras.layers.Dense(32,input_shape=(d_in,), activation='relu'),
    # Capa con 5 salidas, activación tanh
    keras.layers.Dense(48, activation='tanh'),
    #la ultima capa si o si tiene que tener tantas salidas como clases, y softmax 
    keras.layers.Dense(classes, activation='softmax')])

# visualización del modelo inicial (si hay solo 2 variables de entrada)
if d_in==2:
    utils.plot_regresion_logistica2D(modelo,x_train,y_train,title=f"Modelo inicial",detail=0.05)

modelo.compile(
  optimizer=keras.optimizers.SGD(lr=0.001), 
  loss='sparse_categorical_crossentropy', 
  # metricas para ir calculando en cada iteracion o batch 
  # Agregamos el accuracy del modelo
  metrics=['accuracy'], 
)

# Entrenamiento del modelo
history = modelo.fit(x_train,y_train,epochs=200,batch_size=16, validation_data = (x_test,y_test))

# visualización del modelo final (si hay solo 2 variables de entrada)
if d_in==2:
    utils.plot_regresion_logistica2D(modelo,x_train,y_train,title=f"Modelo Final",detail=0.05)
```

```python
# Normalizo las variables de salida
for i in range(d_in):
    x_test[:,i]=(x_test[:,i]-x_test[:,i].mean())/x_test[:,i].std()
```

```python
y_pred = modelo.predict(x_test)
y_pred_labels = np.argmax(y_pred,axis = 1) 
```

```python
plot_frontera_de_decision_2D(modelo, x_train, y_train, x_test=x_test, y_test=y_test, title="",detail=0.05)
```

```python
plot_training_curves(history, acc=True)
```

```python
plot_confusion_matrix(y_test, y_pred_labels)
```

```python
plot_ROC_curve(modelo, x_train, y_train)
```
